```python
#importer les librairies
import pandas as pd
import numpy as np
import matplotlib as plt
from pandas_profiling import ProfileReport
```

```python
#impoter le dataset train
train = pd.read_csv('train.csv')
#renommer la colonne Unnamed en index
train.rename(columns={'Unnamed: 0': 'index'}, inplace =True)
#indexer la colonne index
train.set_index('index',inplace=True)
```

```python
#afficher les 3 premieres lignes du train
train.head(3)
```

```python
#liste des variables catégorielles et numériques

cat = ['credit_hystory','purpose','savings_status','sex','status_matrimonial','other_parties','property_magnitude','other_payment_plans','housing','job','own_telephone','foreign_worker','class']
num = ['checking_status','duration','credit_amount','installment_commitment','residence_since','employment','age','existing_credits','num_dependents']
```

```python
cat
```

```python
num
```

```python
#visualisation des données avec pandas_profiling
prof = ProfileReport(train)
prof.to_file(output_file='rapport.html')
```

```python
#nettoyage des données en retirant les valeurs manquantes et les nulled dans le train

#cherchons tout d'abord les valeurs nulles dans notre dataset train en affichant
train.isnull().sum()

#on remarque qu'il n'y a pas de valeurs nulles mais que faire si il survient des valeurs nulles alors?
#imputation numérique. Ici on remplacera la valeur manquante par la moyenne des valeurs de la colonne
train.fillna(train.mean())
#imputation catégorielle. Ici on remplacera la valeur manquante par la valeur qui se repète le plus
train.fillna(train.mode())
```

```python
#commençcons par le traitement des variables catégorielles
cat
```

```python
#creer de nouvelles features plus comprehensible pour
#les variables catégorielles

train = train.replace(to_replace={'status_matrimonial': {'div/sep':'single','mar/wid':'widow','div/dep/mar':'other'}}, value = None)
train = train.replace(to_replace={'purpose': {'radio/tv':'devices','used car':'car','new car':'car','retrainig':'other','furniture/equipment':'accessory','domestical appliane':'others','repairs':'others'}}, value = None)
train = train.replace(to_replace={'savings_status':{'no known savings':'none','<100':'little','100<=X<500':'moderate','500<=X<1000':'rich','>1000':'quite rich'}}, value = None)
train = train.replace(to_replace={'job': {'unskilled resident':'unskilled','unemp/unskilled no res':'unskilled','high qualif/self emp/mgmt':'high qualif'}}, value = None)
train = train.replace(to_replace={'credit_history': {'critical/other existing credit':'critical'}}, value = None)
```

```python
train
```

```python
#traitement des variables numériques
#features plus comprehensible

train = train.replace(to_replace={'checking_status': {'no checking':0,'<0':0,'0<=X<200':100,'>=200':300}}, value = None)
train = train.replace(to_replace={'employment':{'<1':0,'1<=X<4':4,'4<=X<7':7,'>=7':8}}, value = None)
```
