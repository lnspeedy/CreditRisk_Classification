# 1- Importation des librairies pour tester les différents modèles

```python
import numpy as np
import pandas as pd
import seaborn as sns
from lazypredict.Supervised import LazyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import sklearn.metrics as metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import os
```

```python
train = pd.read_csv('../data/train_final.csv')
```

```python
train.set_index('id', inplace = True)
```

```python
train
```

# 2- Determination de la target value et split du dataset

```python
y = train['class']
del train['class']
X = train 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 40)
```

```python
y_train.shape
```

```python
X_test.shape
```

# 3- Application de la Regression Logistique  

```python
log_reg = LogisticRegression(random_state=10, solver = 'lbfgs')
log_reg.fit(X_train, y_train)
```

```python
log_reg.predict(X_train)
y_pred = log_reg.predict(X_train)
log_reg.predict_proba(X_train)
```

# 4- Précision du test et du train

```python
print('The training Accuracy is ',log_reg.score(X_train, y_train))

print('The testing Accuracy is ',log_reg.score(X_test, y_test))
```
